---------------------------------------------------------------------------

## Step 5: <span style="color:green">Create Quartiles</span> 

```python
Q1 = df_raw.quantile(0.25)
Q3 = df_raw.quantile(0.75)

IQR = Q3 - Q1

print(IQR)
```

---------------------------------------------------------------------------

## Step 6: <span style="color:green">Identify Outliers</span> 

```python
print('There are ' + str(len(df_raw)) + ' entries in the original dataframe.')
```

```python
lower_bound = round(Q1 - (1.5*IQR),2)
upper_bound = round(Q3 + (1.5*IQR),2)


df_no_outliers = df_raw[~((df_raw < (lower_bound))|(df_raw > (upper_bound))).any(axis=1)]
df_outliers = df_raw[((df_raw < (lower_bound))|(df_raw > (upper_bound))).any(axis=1)]

print('There are ' + str(len(df_no_outliers)) + ' entries once we have removed the outliers.')                       
```

```python
print('This means that the proportion of outliers that exist in the dataset is ' + str(round(len(df_outliers) / len(df_raw),2)))
```

---------------------------------------------------------------------------

## Step 7: <span style="color:green">Create a Boxplot without Outliers</span> 

```python
df_no_outliers_failure = df_no_outliers[df_no_outliers['PUMP FAILURE (1 or 0)']==1]
print('When the outliers are removed, there are ' + str(len(df_no_outliers_failure)) + ' instances of pump failure.')
```

```python
df_no_outliers_no_failure = df_no_outliers[df_no_outliers['PUMP FAILURE (1 or 0)']==0]

plt.figure(figsize=(20,5))
sns.boxplot(data = df_no_outliers_no_failure)
plt.title('Data With Outliers Removed and No Pump Failure')
```

### It is not always the right decision to remove outliers. For this problem, the outliers are necessary as they are the primary indicator for when pump failure has occurred. It is important to have an understanding of the subject matter and problem we are trying to solve, otherwise we might have discarded or overlooked this valuable information.

## Step 13: <span style="color:purple">Create a Multivariate Regression Model Using OLS Regression</span> 

```python
dataframe_raw.head()
```

```python
X = dataframe_raw.iloc[:,2:-1]
Y = dataframe_raw.iloc[:,-1]

X = sm.add_constant(X)

model = sm.OLS(Y,X)

results = model.fit()

print(results.summary())
```

```python
X = dataframe_stdev.iloc[:,2:-1]
Y = dataframe_stdev.iloc[:,-1]

X = sm.add_constant(X)

model = sm.OLS(Y,X)

results = model.fit()

print(results.summary())
```

### In a linear regression model, r-squared idicates how close the data are to the fitted regression line. The standard deviation model fits better since the r-squared is 0.78 as opposed to 0.36 for the raw data.  

---------------------------------------------------------------------------

## Step 14: <span style="color:purple">Validate Predictions</span> 

```python
predictions = results.predict(X) 
actual = df_raw['PUMP FAILURE (1 or 0)']
```

```python
fig, ax1 = plt.subplots(figsize=(20,5))

ax1.set_xlabel('Timestamp')
ax1.set_ylabel('ACTUAL PUMP FAILURE (1 or 0)', color='r')
ax1.plot(actual, color='red', label = 'Actual')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.set_ylabel('PREDICTED PUMP FAILURE (1 or 0)', color='blue')  
ax2.plot(predictions, color='b', label='Predictions')
ax2.tick_params(axis='y', labelcolor='blue')



plt.title('Actual vs. Predictions for Pump Failure')
fig.tight_layout()
```
